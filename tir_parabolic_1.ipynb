```python
var('x0', 'y0', 'y1', 'v0', 'a0', 'angle', 't'); g = 9.8;
```

```python
XMIN = -9; XMAX = -XMIN; YMIN = XMIN; YMAX = -XMIN;
```

```python
a(t) = (g*cos(pi*3/2), g*sin(pi*3/2)); a
```

```python
a(1)
```

```python
v(v0, angle, t) = (integral(a[0],t) + v0*cos(angle), integral(a[1],t) + v0*sin(angle)); v
```

```python
p(x0, y0, v0, angle, t) = (v[0](v0, angle, t).integral(t) + x0, v[1](v0, angle, t).integral(t) + y0); p
```
